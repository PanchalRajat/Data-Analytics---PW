### 5. A company has set the following commission structure for its sales team. If a salesperson achieves sales of ₹45,000, use the VLOOKUP function to determine the commission rate applicable.

### Answer:
--> **Formula:** `=VLOOKUP(45000,A2:B5,2,TRUE)`

--> **Explanation:** VLOOKUP searches 45000 in the first column (Sales Amount). Since 45000 is not exactly present, we used **TRUE** which means approximate match. So it takes the nearest smaller value (20000) and returns the commission rate from the 2nd column, which is **10%**.
