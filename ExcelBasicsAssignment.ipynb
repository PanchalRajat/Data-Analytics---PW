## 1. Explain the difference between Relative, Absolute, and Mixed Cell Referencing  
You have the following dataset:

| Vehicle       | Price  |
| ------------- | ------ |
| Car           | 500000 |
| Bicycle       | 10500  |
| Auto-Rickshaw | 25000  |

### → Relative Cell Referencing
- Relative referencing changes when we copy a formula to another cell.  
- **Example:** `A1`  
- If we copy the formula down, `A1` becomes `A2`, then `A3`, and so on.  
- This is useful when we want the formula to adjust automatically.

### → Absolute Cell Referencing
- Absolute referencing does **not** change when we copy the formula.  
- **Example:** `$A$1`  
- No matter where we paste the formula, it will always refer to cell `A1`.  
- This is used when a fixed value is needed, like a tax rate or constant value.

### → Mixed Cell Referencing
- Mixed referencing is a combination of relative and absolute referencing.  
- **Examples:** `$A1` or `A$1`  
  - `$A1` → column is fixed, row can change  
  - `A$1` → row is fixed, column can change  
- This is useful when working with tables where only row or column should stay the same.

## 3. Using the data below, write a formula to calculate the average sales of items priced above 100 but less than 300:
| Item    | Price | Sales |
| ------- | ----- | ----- |
| Item A  | 90    | 1000  |
| Item B  | 150   | 1200  |
| Item C  | 250   | 1500  |

 → To calculate the average sales of items priced above 100 but less than 300, use this formula:

```python
=AVERAGEIFS(C2:C4, B2:B4, ">100", B2:B4, "<300")
```
