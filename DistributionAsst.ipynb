### Question 1: Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Hint: Use BINOM.DIST)

-->  
When a fair die is rolled, there are 6 possible outcomes: {1,2,3,4,5,6}.  
The probability of getting a 3 on a single roll is:

p = 1 / 6  

Number of trials (rolls) = n = 30  
Number of times we want a 3 = x = 5  

This situation follows a **Binomial Distribution** because:
- There are fixed trials (30 rolls)
- Each roll has two outcomes: success (getting 3) or failure (not getting 3)
- Probability of success is constant

The binomial probability formula is:

P(X = x) = C(n,x) × p^x × (1 − p)^(n − x)

Substitute values:

P(X = 5) = C(30,5) × (1/6)^5 × (5/6)^25

Using Excel function:

=BINOM.DIST(5, 30, 1/6, FALSE)

This gives approximately:

P(X = 5) ≈ 0.132

So, the probability of rolling a 3 exactly 5 times in 30 rolls is about **0.132 (13.2%)**.

### Question 3: A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and 55 if the data follows a normal distribution?

-->  
The mean (μ) of the dataset is 50 and the standard deviation (σ) is 5.

The values 45 and 55 are one standard deviation below and above the mean:

45 = 50 − 5 = μ − σ  
55 = 50 + 5 = μ + σ  

According to the Empirical Rule (68–95–99.7 rule) for a normal distribution:

Approximately 68% of the data lies within one standard deviation of the mean.

Therefore, about 68% of the values lie between 45 and 55.

### Question 4: What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.

-->  
Standardization, also called the z-score, is a method used to convert data values into a common scale so that different datasets can be easily compared.

The z-score tells us how many standard deviations a data value is away from the mean.

The formula for z-score is:

z = (x − μ) / σ

Where:  
x = individual data value  
μ = mean of the dataset  
σ = standard deviation of the dataset  

Importance of Standardization:  
Standardization is important because it allows comparison between values from different distributions, identifies how extreme a value is, and helps in detecting outliers.

How Standardization Transforms Data:  
When data is standardized, the original dataset is transformed into a new dataset with a mean of 0 and a standard deviation of 1. This makes the data unit-free and easier to analyze.

### Question 5: What is Kurtosis and their type?

-->  
Kurtosis is a statistical measure that describes the shape of a distribution in terms of how peaked or flat it is compared to a normal distribution.

It tells us about the presence of outliers and how heavy or light the tails of a distribution are.

Types of Kurtosis:

1. Mesokurtic:  
This type has a shape similar to a normal distribution. It has moderate peak and medium tails.

2. Leptokurtic:  
This type has a sharper peak and heavier tails than a normal distribution. It indicates more extreme values (outliers).

3. Platykurtic:  
This type has a flatter peak and lighter tails than a normal distribution. It indicates fewer extreme values.

### Question 6: Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

-->  
A uniform distribution is a distribution in which all outcomes have equal probability.

When rolling a fair die, there are six possible outcomes: {1, 2, 3, 4, 5, 6}. Each face of the die has an equal chance of appearing.

The probability of getting any one number is:

1 / 6

Since all outcomes are equally likely and there is no bias toward any particular number, the outcomes follow a uniform distribution.

Therefore, the uniform distribution is a good model for rolling a fair die because each possible result has the same probability.

### Question 7: Use Excel to compute the probability of getting at least 8 successes in 15 trials with success probability 0.5

-->  
This problem follows a binomial distribution because there is a fixed number of trials (15), each trial has two possible outcomes (success or failure), and the probability of success remains constant at 0.5.

To find the probability of getting at least 8 successes, we first calculate the cumulative probability of getting at most 7 successes using the Excel binomial distribution function:

=BINOM.DIST(7,15,0.5,TRUE)

This formula returns the probability of obtaining 0 to 7 successes in 15 trials.

Since "at least 8 successes" means 8 or more, we subtract this value from 1:

=1 - BINOM.DIST(7,15,0.5,TRUE)

Excel returns approximately 0.5.

Therefore, the probability of getting at least 8 successes in 15 trials is about 0.5, or 50%.

### Question 8: How does log transformation help in stabilizing variance and making data more normally distributed?

-->  
Log transformation is a mathematical technique used to reduce skewness and stabilize variance in a dataset.

Many real-world datasets are positively skewed, meaning they have a long right tail. Applying a logarithm compresses large values more than small values, which reduces the effect of extreme values (outliers).

By reducing skewness, the distribution becomes more symmetric and closer to a normal distribution.

Log transformation also stabilizes variance because it makes the spread of data more consistent across different levels of the variable. When variance increases with the mean, taking logarithms helps control this growth.

Therefore, log transformation improves normality, reduces the impact of outliers, and makes data more suitable for statistical analysis.
