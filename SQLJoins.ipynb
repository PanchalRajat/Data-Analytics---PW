### Question 1: Retrieve all customers who have placed at least one order.

-->  
```
SELECT DISTINCT c.*
FROM Customers c
INNER JOIN Orders o
ON c.CustomerID = o.CustomerID;
```

Explanation:  
This query returns only those customers who have matching records in the Orders table. The INNER JOIN ensures that customers without any orders are excluded, while DISTINCT prevents duplicate customer rows when a customer has placed multiple orders.

###Question 4: Display all customers and orders, whether matched or not.

-->
```
SELECT
    c.CustomerID,
    c.CustomerName,
    o.OrderID,
    o.OrderDate,
    o.Amount
FROM Customers c
LEFT JOIN Orders o
ON c.CustomerID = o.CustomerID

UNION

SELECT
    c.CustomerID,
    c.CustomerName,
    o.OrderID,
    o.OrderDate,
    o.Amount
FROM Customers c
RIGHT JOIN Orders o
ON c.CustomerID = o.CustomerID;
```

Explanation:
This query combines the results of a LEFT JOIN and a RIGHT JOIN using UNION. The LEFT JOIN returns all customers with matching orders, while the RIGHT JOIN returns all orders with matching customers. Using UNION ensures that all customers and all orders are displayed, whether or not they have matching records.

### Question 6: Retrieve customers who made payments but did not place any orders.

-->

```
SELECT DISTINCT c.*
FROM Customers c
INNER JOIN Payments p
ON c.CustomerID = p.CustomerID
LEFT JOIN Orders o
ON c.CustomerID = o.CustomerID
WHERE o.OrderID IS NULL;
```

This query first uses an INNER JOIN to select customers who have made payments. It then applies a LEFT JOIN with the Orders table so that customers without orders will have NULL values. The condition WHERE o.OrderID IS NULL ensures that only customers who made payments but did not place any orders are returned. DISTINCT prevents duplicate customer records.

###Question 8: Show all customers along with order and payment amounts in one table.

-->

```
SELECT
    c.CustomerID,
    c.CustomerName,
    o.OrderID,
    o.Amount AS OrderAmount,
    p.PaymentID,
    p.Amount AS PaymentAmount
FROM Customers c
LEFT JOIN Orders o
ON c.CustomerID = o.CustomerID
LEFT JOIN Payments p
ON c.CustomerID = p.CustomerID;
```

Explanation:
This query uses LEFT JOINs to combine Customers with both Orders and Payments. All customers are included, even if they have no orders or payments. Matching records display their respective amounts, while missing data appears as NULL. Column aliases improve readability by distinguishing order and payment amounts.
