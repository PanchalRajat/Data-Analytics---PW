### Question 1: What is a Common Table Expression (CTE), and how does it improve SQL query readability?

-->  

Answer:  
A Common Table Expression (CTE) is a temporary result set defined within a SQL query using the WITH keyword. It exists only during the execution of that query and is not stored permanently in the database.

CTEs are used to simplify complex queries. Instead of writing long nested subqueries, a CTE allows the query to be broken into smaller logical parts, which makes the query easier to read, understand, and maintain.

Another advantage of CTEs is that the temporary result can be referenced multiple times within the same query, reducing repetition and improving query organization.

Example:

```sql
WITH HighSalaryEmployees AS (
    SELECT emp_id, emp_name, salary
    FROM Employee
    WHERE salary > 50000
)
SELECT *
FROM HighSalaryEmployees;
```

In this example, the CTE stores employees earning more than 50,000, and the main query retrieves the result from that temporary dataset.

### Question 2: Why are some views updatable while others are read-only? Explain with an example.

-->  

Answer:  
In SQL, a view is a virtual table created from the result of a SELECT query. Some views are updatable, meaning we can perform operations like INSERT, UPDATE, or DELETE on them, while others are read-only.

A view is usually updatable when it is based on a single table and does not contain complex operations such as joins, aggregate functions (SUM, AVG, COUNT), GROUP BY, DISTINCT, or subqueries. In such cases, the database can clearly determine how the changes should be applied to the original table.

On the other hand, a view becomes read-only when it is created using multiple tables (joins), aggregate functions, grouping, or other complex expressions. In these situations, the database cannot determine exactly how a change should affect the underlying tables.

Example of an updatable view:

```sql
CREATE VIEW EmployeeView AS
SELECT emp_id, emp_name, salary
FROM Employee;
```

Since this view is based on a single table and contains simple columns, we can update it.

Example:

```sql
UPDATE EmployeeView
SET salary = 60000
WHERE emp_id = 101;
```

Example of a read-only view:

```sql
CREATE VIEW DepartmentSalary AS
SELECT department_id, AVG(salary) AS avg_salary
FROM Employee
GROUP BY department_id;
```

This view uses an aggregate function and GROUP BY, so it is read-only. We cannot update the average salary directly through this view.

Therefore, views are updatable when the database can clearly map the changes to a single underlying table, but they become read-only when the query logic is complex.

### Question 3: What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?

-->  

Answer:  
A stored procedure is a precompiled set of SQL statements that is stored in the database and can be executed whenever needed. Instead of writing the same SQL queries repeatedly, the stored procedure can be called with a single command.

One major advantage of stored procedures is code reusability. Since the logic is stored in the database, the same procedure can be used multiple times by different applications or users without rewriting the query.

Another advantage is better performance. Stored procedures are compiled and optimized by the database engine, which can make execution faster compared to repeatedly sending raw SQL queries from an application.

Stored procedures also improve security. Users can be given permission to execute the procedure without giving them direct access to the underlying tables. This helps protect sensitive data.

They also make maintenance easier. If the query logic needs to change, it only needs to be updated inside the stored procedure rather than modifying the query in multiple places.

Example:

```sql
CREATE PROCEDURE GetAllEmployees
AS
SELECT * FROM Employee;
```

The procedure can then be executed using:

```sql
CALL GetAllEmployees;
```

This approach reduces repetition, improves performance, enhances security, and makes database management more organized.

### Question 4: What is the purpose of triggers in a database? Mention one use case where a trigger is essential.

-->  

Answer:  
A trigger is a special type of stored program in a database that automatically executes when a specific event occurs on a table. These events usually include operations such as INSERT, UPDATE, or DELETE.

The main purpose of triggers is to enforce business rules, maintain data integrity, and automatically perform certain actions whenever data changes. Since triggers run automatically, they help ensure that important operations happen consistently without requiring manual execution of additional queries.

Triggers are often used for tasks such as logging changes to a table, validating data before it is inserted, updating related tables automatically, or maintaining audit records.

One important use case is maintaining an audit log. For example, when a record in an employee table is updated, a trigger can automatically insert a record into an audit table that stores the old and new values along with the time of the change. This helps track modifications and improves data accountability.

Example:

```sql
CREATE TRIGGER log_employee_update
AFTER UPDATE ON Employee
FOR EACH ROW
INSERT INTO Employee_Audit(emp_id, change_date)
VALUES (OLD.emp_id, NOW());
```

In this example, whenever an employee record is updated, the trigger automatically records the change in the audit table.

### Question 5: Explain the need for data modelling and normalization when designing a database.

-->  

Answer:  
Data modelling and normalization are important steps in database design because they help organize data efficiently and ensure that the database structure is logical, consistent, and easy to maintain.

Data modelling is the process of defining how data will be structured, stored, and related within a database. It involves identifying entities (such as students, employees, or orders), their attributes (such as name, salary, or date), and the relationships between them. Proper data modelling helps developers understand how different pieces of data interact and ensures that the database supports the requirements of the system.

Normalization is a technique used to reduce data redundancy and improve data integrity by organizing data into multiple related tables. Instead of storing repeated information in a single table, normalization divides the data into smaller tables and connects them using keys. This prevents duplication, reduces storage waste, and makes updates more consistent.

For example, instead of storing department information repeatedly for every employee in one table, normalization allows the department details to be stored in a separate Department table and linked using a department_id.

By applying proper data modelling and normalization, databases become easier to manage, data remains consistent, and problems such as duplicate data, update anomalies, and inconsistent records can be avoided.

### Question 6: Write a CTE to calculate the total revenue for each product (Revenue = Price × Quantity), and return only products where revenue > 3000.

-->  

```sql
WITH ProductRevenue AS (
    SELECT
        p.ProductID,
        p.ProductName,
        p.Price,
        s.Quantity,
        (p.Price * s.Quantity) AS Revenue
    FROM Products p
    JOIN Sales s
    ON p.ProductID = s.ProductID
)
SELECT ProductID, ProductName, Revenue
FROM ProductRevenue
WHERE Revenue > 3000;
```

Answer:  
This query uses a Common Table Expression (CTE) named **ProductRevenue** to calculate the revenue for each product by multiplying the product price with the quantity sold. The CTE first joins the **Products** and **Sales** tables using `ProductID`. It then computes the revenue for each product. The outer query retrieves only those products whose revenue is greater than **3000**.

### Question 7: Create a view named vw_CategorySummary that shows: Category, TotalProducts, AveragePrice.

-->  

```sql
CREATE VIEW vw_CategorySummary AS
SELECT
    Category,
    COUNT(*) AS TotalProducts,
    AVG(Price) AS AveragePrice
FROM Products
GROUP BY Category;
```

Answer:  
This query creates a view named **vw_CategorySummary**. The view groups the products by their **Category** and calculates two aggregated values: the total number of products using `COUNT(*)` and the average price of products using `AVG(Price)`. The `GROUP BY` clause ensures that these calculations are performed separately for each category.

### Question 8: Create an updatable view containing ProductID, ProductName, and Price. Then update the price of ProductID = 1 using the view.

-->  

```sql
CREATE VIEW vw_ProductDetails AS
SELECT
    ProductID,
    ProductName,
    Price
FROM Products;
```

```sql
UPDATE vw_ProductDetails
SET Price = 1500
WHERE ProductID = 1;
```

Answer:  
This query first creates an updatable view named **vw_ProductDetails** that contains the columns ProductID, ProductName, and Price from the **Products** table. Since the view is based on a single table and does not contain joins, aggregates, or grouping, it is updatable. The second query updates the price of the product with **ProductID = 1** through the view, and the change is automatically reflected in the underlying **Products** table.

### Question 9: Create a stored procedure that accepts a category name and returns all products belonging to that category.

-->  

```sql
CREATE PROCEDURE GetProductsByCategory(IN category_name VARCHAR(50))
BEGIN
    SELECT *
    FROM Products
    WHERE Category = category_name;
END;
```

Answer:  
This query creates a stored procedure named **GetProductsByCategory**. The procedure accepts a parameter called `category_name`, which represents the category to search for. When the procedure is executed, it retrieves all products from the **Products** table that belong to the specified category. This allows the same query logic to be reused easily for different categories.

### Question 10: Create an AFTER DELETE trigger on the Products table that archives deleted product rows into a new table ProductArchive. The archive should store ProductID, ProductName, Category, Price, and DeletedAt timestamp.

-->  

```sql
CREATE TABLE ProductArchive (
    ProductID INT,
    ProductName VARCHAR(100),
    Category VARCHAR(50),
    Price DECIMAL(10,2),
    DeletedAt TIMESTAMP
);
```

```sql
CREATE TRIGGER trg_after_delete_products
AFTER DELETE ON Products
FOR EACH ROW
BEGIN
    INSERT INTO ProductArchive(ProductID, ProductName, Category, Price, DeletedAt)
    VALUES (OLD.ProductID, OLD.ProductName, OLD.Category, OLD.Price, NOW());
END;
```

Answer:  
This solution first creates a table named **ProductArchive** to store information about deleted products. The trigger **trg_after_delete_products** is then created on the **Products** table. Whenever a row is deleted from the Products table, the trigger automatically inserts the deleted product details into the **ProductArchive** table along with the current timestamp using `NOW()`. This ensures that deleted product records are archived for tracking or auditing purposes.
