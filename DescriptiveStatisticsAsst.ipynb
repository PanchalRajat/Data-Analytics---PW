### 1. Understanding Central Tendency (Easy)

A bakery tracks the daily sales of muffins (in dozens) over a week:  
[10, 12, 11, 15, 14, 13, 12].  
What is the most representative value of their weekly sales, and why?

### Answer:
--> The most representative value of weekly sales is the **mean (average)**.

--> When we add all values and divide by total days:  
(10 + 12 + 11 + 15 + 14 + 13 + 12) ÷ 7 = **12.4**

--> Since the sales values are close to each other and there are no extreme values, the **mean** best represents the typical daily muffin sales.

### 2. Mean in Real Life (Easy)

A teacher records the marks of her students in a short quiz:  
[12, 15, 14, 16, 18, 20, 19].  
What is the mean score, and what does it tell us about the class’s performance?

### Answer:
--> The **mean score** is found by adding all marks and dividing by the total number of students.

--> Mean = (12 + 15 + 14 + 16 + 18 + 20 + 19) ÷ 7 = **16.28 (approx.)**

--> This tells us that the **average performance of the class is around 16 marks**, which means most students performed well in the quiz.

### 4. Median in Real Life (Medium)

A car dealer notes the prices of used cars:  
[$8,000, $9,500, $10,200, $11,000, $50,000].  
Why is the median a better measure than the mean in this case? Calculate the median.

### Answer:
--> First, arrange the prices in ascending order (they are already arranged).

--> The **median** is the middle value of the data.
Since there are 5 values, the median is the **3rd value**.

--> **Median = $10,200**

--> The median is a better measure than the mean because there is an **extreme value ($50,000)**.  
This very high price increases the mean and gives a misleading average.

--> The median shows the **typical car price**, which better represents most cars in this dataset.

### 5. Dispersion Introduction (Medium)

A student times how long it takes to finish a puzzle each day:  
[25, 30, 27, 35, 40].  
What does the range tell us about the variation in the student’s puzzle-solving time?

### Answer:
--> The **range** is found by subtracting the smallest value from the largest value.

--> Range = 40 − 25 = **15 minutes**

--> This tells us that the student’s puzzle-solving time **varies by 15 minutes** from the fastest day to the slowest day.

--> A larger range means more variation, so the student’s time is **not very consistent**.

### 6. Range in Action (Medium)

A farmer records the weekly weight of harvested apples (kg):  
[100, 105, 98, 110, 120].

--> Knowing the range helps in **planning packaging size and storage**, as the farmer can prepare for weeks with higher or lower harvest.

Two delivery companies track delivery delays (in minutes).

--> A **lower variance** means delivery times are closer to the average.

--> Since Company A has a smaller variance (6) compared to Company B (15), its delivery delays are **more stable and predictable**.

### 9. Combining Measures (Hard)

A family records their monthly electricity usage (in kWh):  
[400, 420, 390, 450, 410].  
Find the mean and standard deviation. What do these values together tell you about the family’s energy use pattern?

### Answer:
--> **Mean (average)** is calculated by adding all values and dividing by total months.

--> Mean = (400 + 420 + 390 + 450 + 410) ÷ 5 = **414 kWh**

--> **Standard deviation** measures how much the usage varies from the average.
For this data, the standard deviation is **about 20 kWh (approx.)**.

--> **Interpretation:**
- The mean shows that the family uses around **414 kWh per month on average**.
- The low standard deviation means the usage does **not change much each month**.

--> Together, these values tell us that the family’s electricity usage is **fairly consistent and stable**.

### 10. Practical Application (Hard)

A basketball player’s points in 8 games are recorded:  
[15, 18, 20, 22, 25, 17, 19, 21].  
Find the mean, median, mode, range, and standard deviation. What insights can these measures provide about the player’s scoring performance?

### Answer:
--> **Mean (average):**  
(15 + 18 + 20 + 22 + 25 + 17 + 19 + 21) ÷ 8 = **19.6 points (approx.)**

--> **Median:**  
After sorting: [15, 17, 18, 19, 20, 21, 22, 25]  
Median = (19 + 20) ÷ 2 = **19.5 points**

--> **Mode:**  
There is **no mode** because all values occur only once.

--> **Range:**  
25 − 15 = **10 points**

--> **Standard Deviation:**  
Approximately **3 points**.

--> **Insights:**
- The mean and median are very close, showing **consistent scoring**.
- No mode means the player does not rely on one repeated score.
- The small range and low standard deviation indicate **stable performance** with no extreme ups or downs.
