###Question 1: Create a ranking of students based on score (highest first).

-->

```
SELECT
    student_id,
    name,
    course,
    score,
    RANK() OVER (ORDER BY score DESC) AS student_rank
FROM student_performance;
```
Result Data:
https://1drv.ms/x/c/e1a60e1ac5d055b8/IQAn4jXdn0DBSY7AEWcUO05sAS7rzAg-vbSeoiJxSwY63oU?e=fe4tPz

Explanation:
This query uses the RANK() window function to assign a ranking to students based on their score in descending order. The highest score receives rank 1. If two students have the same score, they receive the same rank, and the next rank is skipped.

###Question 10: For each course, find the highest score using FIRST_VALUE().

-->

```
SELECT
    student_id,
    name,
    course,
    score,
    FIRST_VALUE(score) OVER (
        PARTITION BY course
        ORDER BY score DESC
    ) AS highest_score_in_course
FROM student_performance;
```

Result Data:

Explanation:
This query uses the FIRST_VALUE() window function with PARTITION BY course to evaluate each course separately. The scores are ordered in descending order, so FIRST_VALUE() returns the highest score within each course.
