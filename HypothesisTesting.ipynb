### Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?

-->  
The null hypothesis (H₀) is a statement that assumes there is no effect, no difference, or no relationship between variables in a population.

It represents the default or initial assumption that any observed difference in data is due to random chance.

The null hypothesis is important because it provides a clear basis for statistical testing. Hypothesis testing is designed to evaluate whether there is enough evidence to reject the null hypothesis.

By testing H₀, researchers can make objective decisions using data rather than assumptions, which helps ensure reliability and scientific validity of conclusions.

### Question 2: What does the significance level (α) represent in hypothesis testing?

-->  
The significance level, denoted by α (alpha), represents the probability of rejecting the null hypothesis when it is actually true.

In other words, α is the risk of making a Type I error, which occurs when a true null hypothesis is incorrectly rejected.

It sets the threshold for deciding whether the observed results are statistically significant. Common values of α are 0.05 or 0.01.

If the p-value obtained from a test is less than or equal to α, the null hypothesis is rejected; otherwise, it is not rejected.

### Question 3: Differentiate between Type I and Type II errors.

-->  
Type I error occurs when the null hypothesis is rejected even though it is true. This error is also known as a false positive and its probability is represented by the significance level α.

Type II error occurs when the null hypothesis is not rejected even though it is false. This error is also known as a false negative and its probability is represented by β.

In simple terms, a Type I error means detecting an effect that does not actually exist, while a Type II error means failing to detect an effect that actually exists.

### Question 4: Explain the difference between a one-tailed and two-tailed test. Give an example of each.

-->  
A one-tailed test is a hypothesis test in which the direction of the effect is specified. It checks whether a parameter is either greater than or less than a certain value, but not both.

Example of a one-tailed test:  
If a company claims that a new machine increases average production above 100 units, the test checks whether the mean production is greater than 100.

A two-tailed test is a hypothesis test in which no specific direction is mentioned. It checks for any significant difference, whether greater than or less than a certain value.

Example of a two-tailed test:  
If we want to test whether the average production is different from 100 units, the test checks for both increase and decrease from 100.

Thus, a one-tailed test looks in only one direction, while a two-tailed test looks in both directions.

### Question 5: A company claims that the average time to resolve a customer complaint is 10 minutes. A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes. At α = 0.05, test the claim.

-->  
The company’s claim is that the average complaint resolution time is 10 minutes, which is taken as the null hypothesis.

Null hypothesis (H₀): μ = 10  
Alternative hypothesis (H₁): μ ≠ 10  

Since the sample size is small (n = 9) and the population standard deviation is not given, a one-sample t-test is used.

Given data:  
Sample mean (x̄) = 12  
Sample standard deviation (s) = 3  
Sample size (n) = 9  

The test statistic is calculated using the t-test formula:

t = (x̄ − μ) / (s / √n)  

t = (12 − 10) / (3 / √9)  
t = 2 / (3 / 3)  
t = 2  

Degrees of freedom = n − 1 = 8  

At α = 0.05 for a two-tailed test and 8 degrees of freedom, the critical t-value is approximately ±2.306.

Since the calculated t-value (2) is less than the critical value (2.306), we fail to reject the null hypothesis.

Therefore, there is not enough statistical evidence at the 5% significance level to reject the company’s claim that the average complaint resolution time is 10 minutes.

### Question 6: When should you use a Z-test instead of a t-test?

-->  
A Z-test should be used instead of a t-test when the population standard deviation (σ) is known and the sample size is large (usually n ≥ 30).

This is because the Z-test uses the standard normal distribution and gives more accurate results when σ is known, whereas the t-test is designed for situations where σ is unknown and estimated from the sample.

### Question 7: The productivity of 6 employees was measured before and after a training program. At α = 0.05, test if the training improved productivity.

-->  
Since the productivity of the same employees is measured before and after training, this is a paired sample case. Therefore, a paired t-test is appropriate.

Let the hypotheses be:

Null hypothesis (H₀): The training did not improve productivity (mean difference = 0)  
Alternative hypothesis (H₁): The training improved productivity (mean difference > 0)

First, we calculate the difference (After − Before) for each employee:

Employee 1: 55 − 50 = 5  
Employee 2: 65 − 60 = 5  
Employee 3: 59 − 58 = 1  
Employee 4: 58 − 55 = 3  
Employee 5: 63 − 62 = 1  
Employee 6: 59 − 56 = 3  

Mean of differences (d̄) = (5 + 5 + 1 + 3 + 1 + 3) / 6 = 3  

Standard deviation of differences (s) ≈ 1.79  
Sample size (n) = 6  

The test statistic is calculated using:

t = d̄ / (s / √n)

t = 3 / (1.79 / √6)  
t ≈ 4.11  

Degrees of freedom = n − 1 = 5  

At α = 0.05 for a one-tailed test and 5 degrees of freedom, the critical t-value is approximately 2.015.

Since the calculated t-value (4.11) is greater than the critical t-value (2.015), we reject the null hypothesis.

Therefore, there is sufficient statistical evidence at the 5% significance level to conclude that the training program improved employee productivity.

### Question 8: A company wants to test if product preference is independent of gender. At α = 0.05, test independence.

-->  
To test whether product preference is independent of gender, a Chi-square test of independence is used because the data is categorical.

Let the hypotheses be:

Null hypothesis (H₀): Product preference is independent of gender.  
Alternative hypothesis (H₁): Product preference is dependent on gender.

The observed frequencies are:

Male–Product A = 30  
Male–Product B = 20  
Female–Product A = 10  
Female–Product B = 40  

Row totals:  
Male = 50, Female = 50  

Column totals:  
Product A = 40, Product B = 60  

Grand total = 100  

The expected frequencies are calculated using:

Expected frequency = (Row total × Column total) / Grand total

Expected values:

Male–Product A = (50 × 40) / 100 = 20  
Male–Product B = (50 × 60) / 100 = 30  
Female–Product A = (50 × 40) / 100 = 20  
Female–Product B = (50 × 60) / 100 = 30  

The Chi-square test statistic is calculated as:

χ² = Σ (Observed − Expected)² / Expected

χ² = (30−20)²/20 + (20−30)²/30 + (10−20)²/20 + (40−30)²/30  
χ² = 5 + 3.33 + 5 + 3.33  
χ² ≈ 16.66  

Degrees of freedom = (rows − 1)(columns − 1)  
= (2 − 1)(2 − 1) = 1  

At α = 0.05 and 1 degree of freedom, the critical Chi-square value is 3.84.

Since the calculated χ² value (16.66) is greater than the critical value (3.84), we reject the null hypothesis.

Therefore, there is sufficient statistical evidence at the 5% significance level to conclude that product preference is dependent on gender.
