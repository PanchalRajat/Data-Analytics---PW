### Question 1: Create a new database named company_db and create a table named employees with the specified columns.

-->  
To create a new database, use the CREATE DATABASE statement:

CREATE DATABASE company_db;

Next, select the database:

USE company_db;

Now, create the employees table using the CREATE TABLE statement:

CREATE TABLE employees (
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    department VARCHAR(50),
    salary INT,
    hire_date DATE
);

This creates the employees table with the required columns and constraints.

### Question 3: Display all employee records sorted by salary (lowest to highest).

-->  
To sort records in ascending order, use the ORDER BY clause with the salary column:

SELECT *
FROM employees
ORDER BY salary ASC;

This query retrieves all employee records and sorts them from the lowest salary to the highest salary.

### Question 5: List all employees in the IT department, ordered by hire date (newest first).

-->  
To filter records by department and sort by hire date in descending order, use:

SELECT *
FROM employees
WHERE department = 'IT'
ORDER BY hire_date DESC;

This query selects only employees from the IT department and displays the most recently hired employees first.

### Question 9: What is the difference between a Primary Key and a Foreign Key?

-->  
A Primary Key is a column (or set of columns) that uniquely identifies each row in a table. It does not allow duplicate values and cannot contain NULL values.

A Foreign Key is a column that creates a link between two tables. It refers to the Primary Key of another table and is used to maintain relationships between tables.

In simple terms, a Primary Key uniquely identifies records within its own table, while a Foreign Key connects records to another table.

### Question 10: What are constraints in SQL and why are they used?

-->  
Constraints in SQL are rules applied to table columns to enforce data integrity and ensure the accuracy and reliability of the data stored in the database.

They are used to restrict the type of data that can be inserted into a table and to maintain consistency. Examples include PRIMARY KEY, FOREIGN KEY, UNIQUE, NOT NULL, and CHECK constraints.

In simple terms, constraints help prevent invalid or incorrect data from being stored in the database.
