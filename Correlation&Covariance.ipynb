### Question 1: Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

-->  
Covariance is a statistical measure that indicates the direction of the relationship between two variables. It shows whether the variables tend to increase together or whether one increases while the other decreases.

A positive covariance means that both variables increase or decrease together, while a negative covariance means that one variable increases as the other decreases.

Correlation, on the other hand, measures both the strength and direction of the relationship between two variables. Unlike covariance, correlation is standardized and always lies between −1 and +1.

In terms of scale, covariance depends on the units of the variables and can take any value, making it difficult to interpret directly. Correlation is unit-free and easier to interpret because it provides a standardized measure of relationship strength.

### Question 2: What does a positive, negative, and zero covariance indicate about the relationship between two variables?

-->  
A positive covariance indicates that the two variables move in the same direction. When one variable increases, the other variable also tends to increase, and when one decreases, the other also tends to decrease.

A negative covariance indicates that the two variables move in opposite directions. When one variable increases, the other variable tends to decrease, and vice versa.

A zero covariance indicates that there is no linear relationship between the two variables. Changes in one variable do not show any consistent pattern of change in the other variable.

### Question 3: Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?

-->  
One major limitation of covariance is that it depends on the units of measurement of the variables. Because of this, the magnitude of covariance has no clear or standardized interpretation.

Covariance only indicates the direction of the relationship between variables, not the strength of the relationship. This makes it difficult to compare relationships across different datasets.

Correlation is preferred because it standardizes covariance by dividing it by the product of the standard deviations of the variables. As a result, correlation is unit-free and always lies between −1 and +1.

This standardized scale makes correlation easier to interpret and compare, which is why it is commonly used in data analysis.

### Question 4: Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation coefficient. When would you prefer to use Spearman’s correlation?

-->  
Pearson’s correlation coefficient measures the strength and direction of the linear relationship between two continuous variables. It assumes that the data is normally distributed and that the relationship between variables is linear.

Spearman’s rank correlation coefficient measures the strength and direction of the monotonic relationship between two variables using their ranks rather than their actual values. It does not require the data to be normally distributed.

Pearson’s correlation is sensitive to outliers and works best when the data follows a linear pattern, whereas Spearman’s correlation is less affected by outliers because it is based on ranks.

Spearman’s correlation is preferred when the data is not normally distributed, contains outliers, or when the relationship between variables is monotonic but not necessarily linear.

### Question 5: If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you infer causation from this value? Why or why not?

-->  
A correlation coefficient of 0.85 indicates a strong positive relationship between variables X and Y. This means that as the value of X increases, the value of Y also tends to increase in a similar manner.

The value 0.85 being close to 1 shows that the relationship between the two variables is strong, but not perfect.

However, correlation does not imply causation. A high correlation does not mean that changes in X cause changes in Y. The relationship may be influenced by a third variable or may be coincidental.

Therefore, causation cannot be inferred from the correlation coefficient alone without further analysis or experimental evidence.

### Question 6: Using the dataset below, calculate the covariance between X and Y.

-->  
The given data is:

X: 2, 4, 6, 8  
Y: 3, 7, 5, 10  

First, calculate the means of X and Y:

Mean of X (x̄) = (2 + 4 + 6 + 8) / 4 = 5  
Mean of Y (ȳ) = (3 + 7 + 5 + 10) / 4 = 6.25  

Now, calculate (Xi − x̄) and (Yi − ȳ) and their products:

For (2,3): (2 − 5)(3 − 6.25) = (−3)(−3.25) = 9.75  
For (4,7): (4 − 5)(7 − 6.25) = (−1)(0.75) = −0.75  
For (6,5): (6 − 5)(5 − 6.25) = (1)(−1.25) = −1.25  
For (8,10): (8 − 5)(10 − 6.25) = (3)(3.75) = 11.25  

Sum of products = 9.75 − 0.75 − 1.25 + 11.25 = 19  

Covariance is calculated using:

Cov(X,Y) = Σ[(Xi − x̄)(Yi − ȳ)] / n  

Cov(X,Y) = 19 / 4 = 4.75  

Since the covariance is positive, it indicates that X and Y tend to increase together.

### Question 7: Compute the Pearson correlation coefficient between variables A and B.

-->  
The given data is:

A: 10, 20, 30, 40, 50  
B: 8, 14, 18, 24, 28  

First, calculate the means of A and B:

Mean of A (Ā) = (10 + 20 + 30 + 40 + 50) / 5 = 30  
Mean of B (B̄) = (8 + 14 + 18 + 24 + 28) / 5 = 18.4  

Next, calculate deviations from the mean and their products:

For (10, 8):  
(10 − 30) = −20, (8 − 18.4) = −10.4 → Product = 208  

For (20, 14):  
(20 − 30) = −10, (14 − 18.4) = −4.4 → Product = 44  

For (30, 18):  
(30 − 30) = 0, (18 − 18.4) = −0.4 → Product = 0  

For (40, 24):  
(40 − 30) = 10, (24 − 18.4) = 5.6 → Product = 56  

For (50, 28):  
(50 − 30) = 20, (28 − 18.4) = 9.6 → Product = 192  

Sum of products = 208 + 44 + 0 + 56 + 192 = 500  

Now calculate the sum of squared deviations:

Σ(A − Ā)² = 400 + 100 + 0 + 100 + 400 = 1000  
Σ(B − B̄)² = 108.16 + 19.36 + 0.16 + 31.36 + 92.16 = 251.2  

The Pearson correlation coefficient formula is:

r = Σ[(Ai − Ā)(Bi − B̄)] / √[Σ(Ai − Ā)² Σ(Bi − B̄)²]

Substituting values:

r = 500 / √(1000 × 251.2)  
r ≈ 0.998  

This value indicates a very strong positive linear relationship between variables A and B.

### Question 8: The following table shows heights (in cm) and weights (in kg) of 5 students. Find the correlation coefficient between Height and Weight.

-->  
The given data is:

Height (X): 150, 160, 165, 170, 180  
Weight (Y): 50, 55, 58, 62, 70  

First, calculate the mean of Height and Weight:

Mean Height (x̄) = (150 + 160 + 165 + 170 + 180) / 5 = 165  
Mean Weight (ȳ) = (50 + 55 + 58 + 62 + 70) / 5 = 59  

Now calculate deviations from the mean and their products:

For (150, 50):  
(150 − 165) = −15, (50 − 59) = −9 → Product = 135  

For (160, 55):  
(160 − 165) = −5, (55 − 59) = −4 → Product = 20  

For (165, 58):  
(165 − 165) = 0, (58 − 59) = −1 → Product = 0  

For (170, 62):  
(170 − 165) = 5, (62 − 59) = 3 → Product = 15  

For (180, 70):  
(180 − 165) = 15, (70 − 59) = 11 → Product = 165  

Sum of products = 135 + 20 + 0 + 15 + 165 = 335  

Next, calculate the sum of squared deviations:

Σ(X − x̄)² = 225 + 25 + 0 + 25 + 225 = 500  
Σ(Y − ȳ)² = 81 + 16 + 1 + 9 + 121 = 228  

The Pearson correlation coefficient formula is:

r = Σ[(Xi − x̄)(Yi − ȳ)] / √[Σ(Xi − x̄)² Σ(Yi − ȳ)²]

Substituting values:

r = 335 / √(500 × 228)  
r ≈ 0.99  

This value indicates a very strong positive correlation between height and weight.

### Question 10: Two investment portfolios have the following returns (%) over 5 years. Compute the covariance and correlation coefficient, and interpret whether the portfolios move together.

-->  
The given data is:

Portfolio A: 8, 10, 12, 9, 11  
Portfolio B: 6, 9, 11, 8, 10  

First, calculate the mean returns:

Mean of Portfolio A (Ā) = (8 + 10 + 12 + 9 + 11) / 5 = 10  
Mean of Portfolio B (B̄) = (6 + 9 + 11 + 8 + 10) / 5 = 8.8  

Now calculate deviations from the mean and their products:

Year 1:  
(8 − 10)(6 − 8.8) = (−2)(−2.8) = 5.6  

Year 2:  
(10 − 10)(9 − 8.8) = (0)(0.2) = 0  

Year 3:  
(12 − 10)(11 − 8.8) = (2)(2.2) = 4.4  

Year 4:  
(9 − 10)(8 − 8.8) = (−1)(−0.8) = 0.8  

Year 5:  
(11 − 10)(10 − 8.8) = (1)(1.2) = 1.2  

Sum of products = 5.6 + 0 + 4.4 + 0.8 + 1.2 = 12  

Covariance is calculated as:

Cov(A,B) = Σ[(Ai − Ā)(Bi − B̄)] / n  

Cov(A,B) = 12 / 5 = 2.4  

Since the covariance is positive, it indicates that the returns of the two portfolios tend to move in the same direction.

Next, calculate the sum of squared deviations:

Σ(A − Ā)² = 4 + 0 + 4 + 1 + 1 = 10  
Σ(B − B̄)² = 7.84 + 0.04 + 4.84 + 0.64 + 1.44 = 14.8  

The Pearson correlation coefficient is calculated using:

r = Σ[(Ai − Ā)(Bi − B̄)] / √[Σ(Ai − Ā)² Σ(Bi − B̄)²]

r = 12 / √(10 × 14.8)  
r ≈ 0.99  

The correlation coefficient is close to +1, indicating a very strong positive linear relationship.

Therefore, both the covariance and correlation show that Portfolio A and Portfolio B move together, meaning when the return of one portfolio increases or decreases, the return of the other portfolio tends to do the same.
