### Question 1: A die is rolled. What is the probability of getting:  
(a) An even number  
(b) A number greater than 4  

-->  
When a die is rolled, there are 6 equally likely outcomes: {1, 2, 3, 4, 5, 6}.  

(a) Even numbers on a die are {2, 4, 6}.  
Number of favorable outcomes = 3  
Total possible outcomes = 6  
Probability = 3 / 6 = 1 / 2  

(b) Numbers greater than 4 are {5, 6}.  
Number of favorable outcomes = 2  
Total possible outcomes = 6  
Probability = 2 / 6 = 1 / 3

### Question 2: In a class of 50 students:  
20 like Mathematics (M)  
15 like Science (S)  
5 like both subjects  
What is the probability that a student chosen at random likes Mathematics or Science?  

-->  
Some students like only Mathematics, some like only Science, and some like both.  
To find how many like Mathematics or Science, we use:  

M + S − (both)  
= 20 + 15 − 5  
= 30 students  

Total students = 50  

Probability = 30 / 50 = 3 / 5  

This means that 3 out of every 5 students like at least one of the two subjects.

### Question 3: A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability that the next ball is also red (without replacement)?  

-->  
Initially, the bag contains 3 red and 2 blue balls, so total balls = 5.  
One red ball is already taken out and not replaced.  

Remaining red balls = 2  
Remaining total balls = 4  

Probability of drawing a red ball again = 2 / 4 = 1 / 2  

This means there is a 50% chance that the next ball will be red.

### Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling? Why?  

-->  
Stratified Sampling should be used.  
In this method, the population is divided into groups (strata) such as boys and girls.  
Then samples are taken from each group separately.  

This ensures that both boys and girls are properly represented in the sample according to their proportion.

### Question 5: The average height of 1000 students = 160 cm. A sample of 100 students shows an average height = 158 cm. Find the sampling error.  

-->  
Sampling error is the difference between the population mean and the sample mean.  

Sampling Error = Population Mean − Sample Mean  
= 160 − 158  
= 2 cm  

This means the sample average is 2 cm lower than the population average.

### Question 6: The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what is the standard error of the mean (SEM)?  

-->  
Standard Error of Mean (SEM) measures how much the sample mean may vary from the population mean.  

Formula: SEM = σ / √n  

SEM = 5000 / √100  
= 5000 / 10  
= ₹500  

So, the standard error of the mean is ₹500.

### Question 8: From a deck of 52 cards, two cards are drawn without replacement. What is the probability that both are Aces?  

-->  
There are 4 aces in a deck of 52 cards.  

Probability of first ace = 4 / 52  
After drawing one ace, remaining aces = 3  
Remaining cards = 51  

Probability of second ace = 3 / 51  

Total Probability = (4 / 52) × (3 / 51)  
= 12 / 2652  
= 1 / 221  

So, the probability that both cards are aces is 1 / 221.

### Question 9: A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the probability that all are non-defective?  

-->  
Defective rate = 2% = 0.02  
Probability that a bulb is non-defective = 1 − 0.02 = 0.98  

Since 5 bulbs are chosen,  

Probability that all are non-defective = (0.98)^5  
≈ 0.9039  

This means there is about a 90.39% chance that all 5 bulbs are good.

### Question 10: Differentiate between discrete and continuous random variables with examples.  

-->  
Discrete Random Variable:  
A discrete random variable takes only specific and countable values.  
Example: Number of students in a class, number of cars in a parking lot.  

Continuous Random Variable:  
A continuous random variable can take any value within a given range.  
Example: Height of a person, weight of a person, temperature.
